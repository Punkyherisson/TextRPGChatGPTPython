On cree la classe personnage

```python
# Classe Personnage
class Personnage:
    def __init__(self, nom, classe, niveau, pv, attaque, defense, inventaire, pieces_or, xp):
        self.nom = nom
        self.classe = classe
        self.niveau = niveau
        self.pv = pv
        self.attaque = attaque
        self.defense = defense
        self.inventaire = inventaire
        self.pieces_or = pieces_or
        self.xp = xp

    @classmethod
    def creer_depuis_excel(cls, nom_personnage):
        df = pd.read_excel("jeu_de_role.xlsx", sheet_name="Personnages")
        personnage_data = df[df["Nom"] == nom_personnage]
        if personnage_data.empty:
            raise ValueError(f"Personnage '{nom_personnage}' non trouvé dans le fichier Excel.")
        
        personnage_data = personnage_data.iloc[0]
        inventaire = eval(personnage_data["Inventaire"])
        return cls(
            personnage_data["Nom"],
            personnage_data["Classe"],
            personnage_data["Niveau"],
            personnage_data["PV"],
            personnage_data["Attaque"],
            personnage_data["Défense"],
            inventaire,
            personnage_data["PiecesOr"],
            personnage_data["XP"]
        )

    def afficher_stats(self):
        print(f"Nom: {self.nom}\nClasse: {self.classe}\nNiveau: {self.niveau}\nPV: {self.pv}\nAttaque: {self.attaque}\nDéfense: {self.defense}\nXP: {self.xp}\nOr: {self.pieces_or}\nInventaire: {self.inventaire}")

    def calculer_attaque(self):
        # Calculer l'attaque avec l'arme la plus puissante
        df_objets = pd.read_excel("jeu_de_role.xlsx", sheet_name="Objets")
        max_bonus_attaque = 0
        for item in self.inventaire:
            item_data = df_objets[df_objets["Nom de l'objet"] == item]
            if not item_data.empty and item_data.iloc[0]["TypeObjet"] == "Arme":
                max_bonus_attaque = max(max_bonus_attaque, item_data.iloc[0]["Points"])
        return self.attaque + max_bonus_attaque

    def gagner_xp_or(self, xp, pieces_or):
        self.xp += xp
        self.pieces_or += pieces_or
        self.mettre_a_jour_dans_excel()

    def mettre_a_jour_dans_excel(self):
        df = pd.read_excel("jeu_de_role.xlsx", sheet_name="Personnages")
        index = df[df["Nom"] == self.nom].index[0]
        df.at[index, "XP"] = self.xp
        df.at[index, "PiecesOr"] = self.pieces_or
        df.at[index, "PV"] = self.pv
        with pd.ExcelWriter("jeu_de_role.xlsx", engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
            df.to_excel(writer, sheet_name="Personnages", index=False)

    def mettre_a_jour_points_de_vie(self):
        # Mettre à jour les points de vie dans le fichier Excel
        df = pd.read_excel("jeu_de_role.xlsx", sheet_name="Personnages")
        index = df[df["Nom"] == self.nom].index[0]
        df.at[index, "PV"] = self.pv
        with pd.ExcelWriter("jeu_de_role.xlsx", engine="openpyxl", mode="a", if_sheet_exists="replace") as writer:
            df.to_excel(writer, sheet_name="Personnages", index=False)
```

```python
# Fonction pour choisir une mission
def choisir_mission(personnage):
    df_quetes = pd.read_excel("jeu_de_role.xlsx", sheet_name="Quetes")
    print(f"\n=== Missions disponibles pour {personnage.nom} ({personnage.classe}) ===")

    # Filtrer les quêtes en fonction de la classe du personnage
    quetes_disponibles = df_quetes[(df_quetes["Accessible"] == personnage.classe) | (df_quetes["Accessible"] == "Tous")]

    if quetes_disponibles.empty:
        print("Aucune mission disponible pour votre classe.")
        return None, 0

    for index, row in quetes_disponibles.iterrows():
        print(f"{index + 1}. {row['NomdeQuete']}: {row['Descriptif']} (Niveau: {row['Niveau']}, Récompense: {row['RecompenseOr']} Or, {row['RecompenseXP']} XP)")

    choix = input("Entrez le numéro de la mission ou 'r' pour revenir: ").lower()
    if choix == 'r':
        return None, 0
    else:
        choix = int(choix) - 1
        if 0 <= choix < len(quetes_disponibles):
            quete_choisie = quetes_disponibles.iloc[choix]
            print(f"Vous avez choisi la mission: {quete_choisie['NomdeQuete']}")
            return quete_choisie, quete_choisie["NombreMonstres"]
        else:
            print("Choix invalide, veuillez réessayer.")
            return None, 0
        
# Fonction de combat mise à jour
def combat(personnage, monstre, nombre_monstres):
    print(f"\n{personnage.nom} rencontre {nombre_monstres} {monstre['Nom']}!\n")
    
    for i in range(nombre_monstres):
        print(f"Combat contre {monstre['Nom']} {i+1}/{nombre_monstres}")
        while personnage.pv > 0 and monstre["PV"] > 0:
            # Tour du personnage
            attaque_personnage = max(0, personnage.calculer_attaque() - monstre["Defense"])
            monstre["PV"] -= attaque_personnage
            print(f"{personnage.nom} attaque {monstre['Nom']} pour {attaque_personnage} dégâts. {monstre['Nom']} a maintenant {monstre['PV']} PV.")
            if monstre["PV"] <= 0:
                print(f"{monstre['Nom']} est vaincu!")
                break

            # Tour du monstre
            attaque_monstre = max(1, monstre["Attaque"] - personnage.defense)
            personnage.pv -= attaque_monstre
            print(f"{monstre['Nom']} attaque {personnage.nom} pour {attaque_monstre} dégâts. {personnage.nom} a maintenant {personnage.pv} PV.")
            if personnage.pv <= 0:
                print(f"{personnage.nom} est vaincu!")
                return False  # Le personnage est vaincu
        
        if personnage.pv > 0:
            # Après chaque combat, demander si le joueur veut continuer ou fuir
            choix = input(f"Il reste {nombre_monstres - i - 1} monstres. Voulez-vous continuer (c) ou fuir (f)? ").lower()
            if choix == 'f':
                print("Vous avez choisi de fuir la mission.")
                return True  # Le personnage fuit la mission
        
        # Reset monstre PV pour le prochain combat
        monstre["PV"] = df_monstres[df_monstres["Nom"] == monstre["Nom"]]["PV"].values[0]

    if personnage.pv > 0:
        # Si le personnage a vaincu tous les monstres, mettre à jour XP et Or
        xp_gagne = monstre["PtsXP"] * nombre_monstres
        or_gagne = monstre["Or"] * nombre_monstres
        # Ajouter l'XP et l'or de la mission aux récompenses obtenues en tuant les monstres
        xp_gagne_total = xp_gagne + quete_choisie["RecompenseXP"]
        or_gagne_total = or_gagne + quete_choisie["RecompenseOr"]
        print(f"{personnage.nom} a vaincu tous les {monstre['Nom']} et termine la mission avec succès!\n"
              f"Vous avez gagné {xp_gagne_total} points d'XP et {or_gagne_total} pièces d'or.")
        personnage.gagner_xp_or(xp_gagne_total, or_gagne_total)
        # Mettre à jour les points de vie dans le fichier Excel
        personnage.mettre_a_jour_points_de_vie()
    else:
        print(f"{personnage.nom} a été vaincu.")

    # Mettre à jour les points de vie dans le fichier Excel
    personnage.mettre_a_jour_points_de_vie()
    
# Initialiser un exemple de personnage et monstre pour le test
personnage_test = Personnage.creer_depuis_excel("Galahad")
df_monstres = pd.read_excel("jeu_de_role.xlsx", sheet_name="Monstres")
monstre_test = df_monstres[df_monstres["Nom"] == "Ogre"].iloc[0].to_dict()

# Choisir une mission pour le personnage et lancer le combat
quete_choisie, nombre_monstres = choisir_mission(personnage_test)
if quete_choisie is not None:
    nom_monstre = quete_choisie["NomMonstre"]
    if nom_monstre:
        monstre = df_monstres[df_monstres["Nom"] == nom_monstre].iloc[0].to_dict()
        combat(personnage_test, monstre, nombre_monstres)
    else:
        print("Cette mission ne nécessite pas de combattre des monstres.")
```

```python
import math
def marchand(personnage):
    while True:
        print("\n=== Marchand ===")
        print("Bienvenue chez le marchand!")
        print("1. Acheter des objets")
        print("2. Vendre des objets")
        print("3. Revenir au menu précédent")

        choix = input("Que souhaitez-vous faire? ").strip()

        if choix == "1":
            acheter_objets(personnage)
        elif choix == "2":
            vendre_objets(personnage)
        elif choix == "3":
            return
        else:
            print("Choix invalide.")

def acheter_objets(personnage):
    print("\n=== Acheter des objets ===")
    df_objets = pd.read_excel("jeu_de_role.xlsx", sheet_name="Objets")
    print("Voici les objets disponibles à l'achat:")
    print(df_objets)

    while True:
        choix = input("Entrez le numéro de l'objet que vous souhaitez acheter ou 'q' pour quitter: ").strip()
        if choix == "q":
            break

        try:
            choix = int(choix) - 1
            objet_choisi = df_objets.iloc[choix]
            prix = objet_choisi["Prix"]
            if personnage.pieces_or >= prix:
                print(f"Vous avez acheté {objet_choisi['Nom de l\'objet']} pour {prix} pièces d'or.")
                personnage.pieces_or -= prix
                personnage.inventaire.append(objet_choisi["Nom de l'objet"])
                print(f"Il vous reste {personnage.pieces_or} pièces d'or.")
            else:
                print("Vous n'avez pas assez d'or pour acheter cet objet.")
        except (ValueError, IndexError):
            print("Choix invalide. Veuillez entrer un numéro valide.")

def vendre_objets(personnage):
    print("\n=== Vendre des objets ===")
    print("Voici les objets dans votre inventaire:")
    print(personnage.inventaire)

    while True:
        choix = input("Entrez le numéro de l'objet que vous souhaitez vendre ou 'q' pour quitter: ").strip()
        if choix == "q":
            break

        try:
            choix = int(choix) - 1
            objet_choisi = personnage.inventaire[choix]
            index_objet = personnage.inventaire.index(objet_choisi)
            prix = math.ceil(df_objets[df_objets["Nom de l'objet"] == objet_choisi]["Prix"] / 2)
            print(f"Vous avez vendu {objet_choisi} pour {prix} pièces d'or.")
            personnage.pieces_or += prix
            del personnage.inventaire[index_objet]
            print(f"Vous avez maintenant {personnage.pieces_or} pièces d'or.")
        except (ValueError, IndexError):
            print("Choix invalide. Veuillez entrer un numéro valide.")

# Création de l'objet Galahad à partir de la feuille Excel
galahad = Personnage.creer_depuis_excel("Galahad")
print(galahad)
# Appel de la fonction du marchand pour Galahad
marchand(galahad)
```
